```python
import pandas as pd
import numpy as np

items_raw = pd.read_csv('/root/shared/sales/items.csv')
train_raw = pd.read_csv('/root/shared/sales/sales_train_v2.csv')
shops_raw = pd.read_csv('/root/shared/sales/shops.csv')
categories_raw = pd.read_csv('/root/shared/sales/item_categories.csv')
```

```python
import datetime
import math
def time2vec(timeStr):
    day = datetime.datetime.strptime(timeStr, '%d.%m.%Y').timetuple().tm_yday
    cos = math.cos(day/365*math.pi*2)
    sin = math.sin(day/365*math.pi*2)
    return np.array([sin, cos])
```

```python
unique_dates_raw = train_raw['date'].unique()
```
