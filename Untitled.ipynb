```python
def remove_elements(string, elements = ['!', '"', '#', '$', '%', '&', '(', ')', '*', '+', ',', '-', '.', '/', ':', ';', '<', '=', '>', '?', '@', '[', '\\', ']', '^', '_', '`', '{', '|', '}', '~', '–', '‘', '…']):
    string = string.lower().strip()
    for i in elements:
        string = string.replace(i, ' ')
    while (string.find('  ') != -1):
        string = string.replace('  ', ' ')
    return string.strip()
```

```python
def time2vec(timeStr):
    day = datetime.datetime.strptime(timeStr, '%d.%m.%Y').timetuple().tm_yday
    cos = math.cos(day/365*math.pi*2)
    sin = math.sin(day/365*math.pi*2)
    return np.array([cos, sin])
```

```python
def dec2vec(num):
    pos = 1 if num>=0 else 0
    b = bin(num if pos else -num)[2:]
    return np.array([pos] + [int(x) for x in '0'*(12-len(b)) + b])
    
```

```python
from keras.layers import Input, Embedding, LSTM, Dense, Concatenate
from keras.models import Model
from keras.preprocessing.text import one_hot
from keras.preprocessing.sequence import pad_sequences
import keras
```

```python
main_input = Input(shape=(main_input_data.astype('float32').shape[1],), dtype='float32', name='main_input')
name_input = Input(shape=(name_input_data.shape[1],), dtype='int32', name='name_input')

emb = Embedding(output_dim=512, input_dim=vocab_size, input_length=max_length)(name_input)
# emb = Embedding(vocab_size, 64, input_length=max_length)(name_input)
lstm_out = LSTM(32)(emb)

auxiliary_output = Dense(13, activation='softmax', name='aux_output')(lstm_out)

x = keras.layers.concatenate([main_input, lstm_out])

x = Dense(64, activation='relu')(x)
x = Dense(64, activation='relu')(x)
x = Dense(64, activation='relu')(x)

# And finally we add the main logistic regression layer
main_output = Dense(13, activation='softmax', name='main_output')(x)
```

```python
model = Model(inputs=[main_input, name_input], outputs=[main_output, auxiliary_output])

model.compile(optimizer='adam', loss='binary_crossentropy', loss_weights=[1., 0.2], metrics=['acc'])

print(model.summary())
```

```python
model.fit([main_input_data, name_input_data], [labels, labels], epochs=50, batch_size=32, validation_split=0.1)
```
