```python
import datetime
import math
def time2vec(timeStr):
    day = datetime.datetime.strptime(timeStr, '%d.%m.%Y').timetuple().tm_yday
    cos = math.cos(day/365*math.pi*2)
    sin = math.sin(day/365*math.pi*2)
    return np.array([sin, cos])
```

```python
def getForPeriod(start, end):
    startId = dates.loc[dates['date'] == start].index[0]
    
#     endId = dates.loc[dates['date'] == end].index[0]
    endId = dates.loc[dates['date'] == end].index[0] -1
    end = dates.loc[endId,'date']
    
    amount = endId-startId+1
    labels = np.zeros((amount,len(items_raw)))
    features = np.zeros((amount,len(items_raw), len(shops_raw) + len(categories_raw) + 2 + 2)) 
    
    for i, v in train_raw.iterrows():
        date = v['date']
        dateIndex = dates.loc[dates['date'] == date].index[0]
        if endId >= dateIndex >= startId:
            labelDateIndex = dateIndex-startId
            labels[labelDateIndex][v['item_id']] = v['item_cnt_day']
            features[labelDateIndex][v['item_id']] = np.concatenate((shop_buyId.iloc[[i]].values.flatten(), categories_1hot.iloc[[items_raw.loc[v['item_id'],'item_category_id']]].values.flatten(), np.array((dates.loc[dateIndex,'sin'], dates.loc[dateIndex,'cos'], v['item_price']/items_raw.loc[v['item_id'], 'maxPrice'], v['item_price']/maxPrice))), axis=0)
    
    p = np.random.permutation(len(items_raw))
    labels = labels.T[p].T
    features = np.transpose(np.transpose(features,(1,0,2))[p], (1,0,2))
    return labels,features
```
